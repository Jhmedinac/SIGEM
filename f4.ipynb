# [dbo].[vw_eventos_completo]

```sql
/****** Object:  View [dbo].[vw_eventos_completo]    Script Date: 17/2/2026 14:03:04 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO


-- =============================================
-- 4. VISTAS PRINCIPALES
-- =============================================

-- Vista Completa de Eventos
CREATE VIEW [dbo].[vw_eventos_completo] AS
SELECT
    e.id_evento,
    e.codigo_evento,
    e.nombre_evento,
    e.descripcion,
    e.objetivo,
    e.fecha_inicio,
    e.fecha_fin,
    e.hora_inicio,
    e.hora_fin,
    e.cupo_maximo,
    e.cupo_actual,
    (e.cupo_maximo - e.cupo_actual) AS cupos_disponibles,
    e.activo,
    -- IDs
    e.id_estado,
    e.id_modalidad,
    e.id_tipo_evento,
    e.id_facilitador,
    e.id_salon,
    e.id_coordinador,
    e.id_gerencia_solicitante,
    e.id_solicitante,
    -- Descripciones
    es.nombre_estado AS estado_desc,
    m.nombre_modalidad AS modalidad_desc,
    te.nombre_tipo AS tipo_evento_desc,
    ISNULL(s.nombre_salon, 'VIRTUAL / POR DEFINIR') AS nombre_salon,
    -- Facilitador
    CASE
        WHEN f.tipo_facilitador = 'INTERNO' THEN (p_fac.primernombre + ' ' + p_fac.primerapellido)
        ELSE (ISNULL(f.nombres, '') + ' ' + ISNULL(f.apellidos, ''))
    END AS facilitador_nombre
FROM bd.evento e
LEFT JOIN bd.estado_solicitud es ON e.id_estado = es.id_estado
LEFT JOIN bd.modalidad m ON e.id_modalidad = m.id_modalidad
LEFT JOIN bd.tipo_evento te ON e.id_tipo_evento = te.id_tipo_evento
LEFT JOIN bd.salon s ON e.id_salon = s.id_salon
LEFT JOIN bd.facilitador f ON e.id_facilitador = f.id_facilitador
LEFT JOIN bd.persona p_fac ON f.id_persona = p_fac.id_persona;
GO
```

# [bd].[sp_cambiar_estado_evento]

```sql
/****** Object:  StoredProcedure [bd].[sp_cambiar_estado_evento]    Script Date: 17/2/2026 14:03:04 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- =============================================
-- OTROS PROCEDIMIENTOS ALMACENADOS
-- =============================================

CREATE PROCEDURE [bd].[sp_cambiar_estado_evento]
    @id_evento INT,
    @nuevo_estado VARCHAR(50) -- Recibe 'APROBADO', 'CANCELADO', etc.
AS
BEGIN
    SET NOCOUNT ON;
    DECLARE @id_nuevo_estado INT;

    SELECT TOP 1 @id_nuevo_estado = id_estado
    FROM bd.estado_solicitud
    WHERE nombre_estado = @nuevo_estado;

    IF @id_nuevo_estado IS NOT NULL
    BEGIN
        -- Logica de Historial
        INSERT INTO bd.historial_estado_evento (id_evento, id_estado_anterior, id_estado_nuevo, usuario_cambio, motivo_cambio)
        SELECT id_evento, id_estado, @id_nuevo_estado, 'SISTEMA', 'Cambio de estado automatico o manual'
        FROM bd.evento WHERE id_evento = @id_evento;

        UPDATE bd.evento
        SET
            id_estado = @id_nuevo_estado,
            fecha_modificacion = GETDATE()
        WHERE id_evento = @id_evento;
    END
END
GO
```

# [bd].[sp_consultar_disponibilidad_salon]

```sql
/****** Object:  StoredProcedure [bd].[sp_consultar_disponibilidad_salon]    Script Date: 17/2/2026 14:03:04 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Validar Disponibilidad Salon
CREATE PROCEDURE [bd].[sp_consultar_disponibilidad_salon]
    @fecha_inicio DATE,
    @fecha_fin DATE
AS
BEGIN
    SET NOCOUNT ON;

    SELECT
        s.id_salon,
        s.nombre_salon,
        s.ubicacion,
        s.capacidad_maxima,
        CASE
            WHEN EXISTS (
                SELECT 1 FROM bd.evento e
                WHERE e.id_salon = s.id_salon
                AND e.activo = 1
                AND (
                    (e.fecha_inicio BETWEEN @fecha_inicio AND @fecha_fin) OR
                    (e.fecha_fin BETWEEN @fecha_inicio AND @fecha_fin) OR
                    (@fecha_inicio BETWEEN e.fecha_inicio AND e.fecha_fin)
                )
            ) THEN 'OCUPADO'
            ELSE 'DISPONIBLE'
        END AS estado_disponibilidad
    FROM bd.salon s
    WHERE s.activo = 1
    ORDER BY s.nombre_salon;
END;
GO
```

# [bd].[sp_crear_evento]

```sql
/****** Object:  StoredProcedure [bd].[sp_crear_evento]    Script Date: 17/2/2026 14:03:04 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- =============================================
-- 5. PROCEDURES OPTIMIZADOS
-- =============================================

CREATE PROCEDURE [bd].[sp_crear_evento]
    @codigo_evento VARCHAR(50),
    @nombre_evento VARCHAR(255),
    @descripcion VARCHAR(MAX),
    @objetivo VARCHAR(MAX),
    @id_tipo_evento INT,
    @id_modalidad INT,
    @id_coordinador INT,
    @fecha_inicio DATE,
    @fecha_fin DATE,
    @hora_inicio TIME,
    @hora_fin TIME,
    @anio INT,
    @id_salon INT = NULL,
    @cupo_maximo INT,
    @usuario_registro VARCHAR(50),
    @id_evento INT = NULL,
    @id_facilitador INT = NULL,
    @id_gerencia_solicitante VARCHAR(50) = NULL,
    @id_solicitante INT = NULL
AS
BEGIN
    SET NOCOUNT ON;

    -- Default Estado: PENDIENTE
    DECLARE @id_estado_inicial INT;
    SELECT TOP 1 @id_estado_inicial = id_estado FROM bd.estado_solicitud WHERE nombre_estado = 'PENDIENTE';
    IF @id_estado_inicial IS NULL SELECT TOP 1 @id_estado_inicial = id_estado FROM bd.estado_solicitud;

    IF @id_evento IS NULL OR @id_evento = 0
    BEGIN
        INSERT INTO bd.evento (
            codigo_evento, nombre_evento, id_tipo_evento, id_modalidad, id_coordinador, id_estado,
            fecha_inicio, fecha_fin, hora_inicio, hora_fin, anio, cupo_maximo, cupo_actual,
            fecha_solicitud, activo, descripcion, objetivo, id_salon, usuario_creacion,
            id_facilitador, id_gerencia_solicitante, id_solicitante
        )
        VALUES (
            @codigo_evento, @nombre_evento, @id_tipo_evento, @id_modalidad, @id_coordinador, @id_estado_inicial,
            @fecha_inicio, @fecha_fin, @hora_inicio, @hora_fin, @anio, @cupo_maximo, 0,
            GETDATE(), 1, @descripcion, @objetivo, @id_salon, @usuario_registro,
            @id_facilitador, @id_gerencia_solicitante, @id_solicitante
        );
    END
    ELSE
    BEGIN
        UPDATE bd.evento
        SET
            codigo_evento = @codigo_evento,
            nombre_evento = @nombre_evento,
            descripcion = @descripcion,
            objetivo = @objetivo,
            id_tipo_evento = @id_tipo_evento,
            id_modalidad = @id_modalidad,
            id_coordinador = @id_coordinador,
            fecha_inicio = @fecha_inicio,
            fecha_fin = @fecha_fin,
            hora_inicio = @hora_inicio,
            hora_fin = @hora_fin,
            anio = @anio,
            id_salon = @id_salon,
            cupo_maximo = @cupo_maximo,
            id_facilitador = @id_facilitador,
            id_gerencia_solicitante = @id_gerencia_solicitante,
            id_solicitante = @id_solicitante,
            fecha_modificacion = GETDATE(),
            usuario_modificacion = @usuario_registro
        WHERE id_evento = @id_evento;
    END
END
GO
```

# [bd].[sp_dashboard_estadisticas]

```sql
/****** Object:  StoredProcedure [bd].[sp_dashboard_estadisticas]    Script Date: 17/2/2026 14:03:04 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [bd].[sp_dashboard_estadisticas]
    @p_anio INT
AS
BEGIN
    SET NOCOUNT ON;

    -- Eventos por estado
    SELECT es.nombre_estado, COUNT(*) AS total_eventos
    FROM bd.evento e
    INNER JOIN bd.estado_solicitud es ON e.id_estado = es.id_estado
    WHERE e.anio = @p_anio
    GROUP BY es.nombre_estado;

    -- Resumen global
    SELECT
        COUNT(DISTINCT part.id_persona) AS total_participantes_unicos,
        COUNT(*) AS total_inscripciones,
        SUM(CASE WHEN part.asistencia = 1 THEN 1 ELSE 0 END) AS total_asistencias
    FROM bd.participante part
    INNER JOIN bd.evento e ON part.id_evento = e.id_evento
    WHERE e.anio = @p_anio;
END
GO
```

# [bd].[sp_estadisticas_eventos_anio]

```sql
/****** Object:  StoredProcedure [bd].[sp_estadisticas_eventos_anio]    Script Date: 17/2/2026 14:03:04 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [bd].[sp_estadisticas_eventos_anio]
    @anio INT
AS
BEGIN
    SET NOCOUNT ON;

    SELECT
        te.nombre_tipo AS tipo_evento,
        COUNT(e.id_evento) AS total_eventos,
        SUM(e.cupo_actual) AS total_participantes,
        AVG(CAST(e.cupo_actual AS FLOAT)) AS promedio_participantes,
        SUM(e.duracion_horas) AS total_horas_capacitacion
    FROM bd.evento e
    INNER JOIN bd.tipo_evento te ON e.id_tipo_evento = te.id_tipo_evento
    WHERE e.anio = @anio AND e.activo = 1
    GROUP BY te.nombre_tipo
    ORDER BY total_eventos DESC;
END
GO
```

# [bd].[sp_inscribir_participante]

```sql
/****** Object:  StoredProcedure [bd].[sp_inscribir_participante]    Script Date: 17/2/2026 14:03:04 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Inscribir Participante
CREATE PROCEDURE [bd].[sp_inscribir_participante]
    @p_id_evento INT,
    @p_id_persona INT,
    @p_mensaje VARCHAR(255) OUTPUT
AS
BEGIN
    SET NOCOUNT ON;
    DECLARE @v_cupo_actual INT;
    DECLARE @v_cupo_maximo INT;
    DECLARE @v_existe INT;

    BEGIN TRY
        BEGIN TRANSACTION;

        -- Verificar si ya existe
        SELECT @v_existe = COUNT(*)
        FROM bd.participante
        WHERE id_evento = @p_id_evento AND id_persona = @p_id_persona;

        IF @v_existe > 0
        BEGIN
            SET @p_mensaje = 'El participante ya esta inscrito.';
            ROLLBACK TRANSACTION;
            RETURN;
        END

        -- Verificar cupos
        SELECT @v_cupo_actual = cupo_actual, @v_cupo_maximo = cupo_maximo
        FROM bd.evento WHERE id_evento = @p_id_evento;

        IF @v_cupo_actual >= @v_cupo_maximo
        BEGIN
            SET @p_mensaje = 'No hay cupos disponibles.';
            ROLLBACK TRANSACTION;
            RETURN;
        END

        -- Inscribir
        INSERT INTO bd.participante (id_evento, id_persona, estado_participacion)
        VALUES (@p_id_evento, @p_id_persona, 'INSCRITO');

        -- Actualizar contador
        UPDATE bd.evento SET cupo_actual = cupo_actual + 1 WHERE id_evento = @p_id_evento;

        SET @p_mensaje = 'Participante inscrito exitosamente';
        COMMIT TRANSACTION;
    END TRY
    BEGIN CATCH
        IF @@TRANCOUNT > 0 ROLLBACK TRANSACTION;
        SET @p_mensaje = 'Error SQL: ' + ERROR_MESSAGE();
    END CATCH
END;
GO
```

# [bd].[sp_login_usuario]

```sql
/****** Object:  StoredProcedure [bd].[sp_login_usuario]    Script Date: 17/2/2026 14:03:04 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Procedimiento de Login
CREATE PROCEDURE [bd].[sp_login_usuario]
    @p_username VARCHAR(100)
AS
BEGIN
    SET NOCOUNT ON;

    SELECT
        u.id_usuario,
        u.username,
        u.password_hash,
        r.nombre_rol,
        p.id_persona,
        p.primernombre,
        p.primerapellido,
        p.email,
        u.bloqueado,
        u.activo
    FROM bd.usuario u
    INNER JOIN bd.rol r ON u.id_rol = r.id_rol
    INNER JOIN bd.persona p ON u.id_persona = p.id_persona
    WHERE u.username = @p_username;
END
GO
```

# [bd].[sp_registrar_evaluacion]

```sql
/****** Object:  StoredProcedure [bd].[sp_registrar_evaluacion]    Script Date: 17/2/2026 14:03:04 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

-- Registrar Evaluacion
CREATE PROCEDURE [bd].[sp_registrar_evaluacion]
    @id_participante INT,
    @calificacion_contenido DECIMAL(3,2),
    @calificacion_facilitador DECIMAL(3,2),
    @calificacion_organizacion DECIMAL(3,2),
    @calificacion_general DECIMAL(3,2),
    @comentarios NVARCHAR(MAX) = NULL,
    @sugerencias NVARCHAR(MAX) = NULL,
    @aplicabilidad_trabajo VARCHAR(10) = NULL,
    @recomendaria BIT = NULL
AS
BEGIN
    SET NOCOUNT ON;

    -- Insertar o Update
    MERGE bd.evaluacion AS target
    USING (SELECT @id_participante AS id_participante) AS source
    ON (target.id_participante = source.id_participante)
    WHEN MATCHED THEN
        UPDATE SET
            calificacion_contenido = @calificacion_contenido,
            calificacion_facilitador = @calificacion_facilitador,
            calificacion_organizacion = @calificacion_organizacion,
            calificacion_general = @calificacion_general,
            comentarios = @comentarios,
            sugerencias = @sugerencias,
            aplicabilidad_trabajo = @aplicabilidad_trabajo,
            recomendaria = @recomendaria,
            fecha_evaluacion = GETDATE()
    WHEN NOT MATCHED THEN
        INSERT (id_participante, calificacion_contenido, calificacion_facilitador, calificacion_organizacion, calificacion_general, comentarios, sugerencias, aplicabilidad_trabajo, recomendaria)
        VALUES (@id_participante, @calificacion_contenido, @calificacion_facilitador, @calificacion_organizacion, @calificacion_general, @comentarios, @sugerencias, @aplicabilidad_trabajo, @recomendaria);
END;
GO
```
